```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df_random = pd.read_csv("sort_data/selection_sort_generate_random_list.csv")
# df_random1 = pd.read_csv("sort_data/quick_sort_generate_random_list.csv")

# df_almost = pd.read_csv("sort_data/quick_sort_generate_almost_sorted_list.csv")
# df_reverse = pd.read_csv("sort_data/merge_sort_generate_almost_reverse_sorted_list.csv")
```

```python
group1 = df_random.groupby('List Size')['Time Taken (s)'].mean()
# group2 = df_random1.groupby('List Size')['Time Taken (s)'].mean()
# group3 = df_reverse.groupby('List Size')['Time Taken (s)'].mean()

# groups = [group1, group2, group3]

# df_grouped = pd.concat(groups).groupby('List Size').mean()
# df_grouped.reset_index()
```

```python
df = group1
df = df.reset_index()
theoretical_df = pd.DataFrame()
theoretical_df['List Size'] = [i for i in range(0, df["List Size"].iloc[-1] + 5, 5)]
```

```python
# Compute theoretical times and add as a new column
linearithmic_normalizer = df['Time Taken (s)'].iloc[1] / (df["List Size"].iloc[1] * np.log2(df["List Size"].iloc[1]))
theoretical_df["Linearithmic Time (s)"] = linearithmic_normalizer * theoretical_df['List Size'] * np.log2(theoretical_df['List Size'])

# Compute theoretical times and add as a new column
linear_normalizer = df["Time Taken (s)"].iloc[1] / (df["List Size"].iloc[1])
theoretical_df["Linear Time (s)"] = linear_normalizer * theoretical_df['List Size']

# Compute theoretical times and add as a new column
poloynomial_normalizer = df["Time Taken (s)"].iloc[1] / (df["List Size"].iloc[1] ** 2)
theoretical_df["Polynomial Time (s)"] = poloynomial_normalizer * (theoretical_df["List Size"] ** 2)

```

```python

# Plot using pandas
plt.figure(figsize=(10, 6))
plt.plot(df["List Size"], df["Time Taken (s)"], 'o-', label="Measured Times")
plt.plot(theoretical_df["List Size"], theoretical_df["Linearithmic Time (s)"], '-', label=r"Theoretical $O(n \log n)$")
plt.plot(theoretical_df["List Size"], theoretical_df["Linear Time (s)"], '-', label=r"Theoretical $O(n)$")
plt.plot(theoretical_df["List Size"], theoretical_df["Polynomial Time (s)"], '-', label=r"Theoretical $O(n^2)$")



plt.xlabel("List Size")
plt.ylabel("Time (seconds)")
plt.title("Performance Comparison: Measured vs Theoretical")
plt.ylim(0,50)
plt.legend()
plt.grid(True)
plt.show()
```
